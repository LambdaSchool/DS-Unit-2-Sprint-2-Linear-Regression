# Model Diagnostics

The purpose of this assigment is introduce you to a new library for linear regression called statmodels which is much better suited for inferential modeling than sklearn. This assignment is also to familiarize yourself with some of most important procedures for improving the interpretability of regression coefficients. You will also perform important statistical tests that will help establish that whether or not important assumptions that safeguard the interpretability of OLS coefficients have been met. 

We will continue to use the Ames Housing Dataset so that you can focus on the techniques and not on cleaning/getting associated with a brand new dataset.

```python
##### Your Code Here #####
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
from scipy import stats
model = LinearRegression()
df = pd.read_csv("https://raw.githubusercontent.com/ryanleeallred/datasets/master/Ames%20Housing%20Data/train.csv")
df = df.replace(np.nan,0)
df.head()
```

```python
sqfteq = pd.concat([df["1stFlrSF"],df["1stFlrSF"]**2,df["2ndFlrSF"],df["2ndFlrSF"]**2,df["1stFlrSF"]+df["2ndFlrSF"]],axis=1)
baths = pd.DataFrame(df["BsmtFullBath"]+df["BsmtHalfBath"]+df["FullBath"]+df["HalfBath"])
baths.columns =["TotalBaths"]
sqfteq.columns = ["1F","2F","1FSquared","2FSquared","1F+2F"]
```

## 1.2 Now using the X variables that you feel like will be the best predictors of y use statsmodel to run the multiple regression between these variables and Y. You don't need to use every X variable in your dataset, in fact it's probably better if you don't. Just pick ones that you have already cleaned that seem the most relevant to house prices.

```python
##### Your Code Here #####
x = pd.concat( [ df[["OverallQual", "GarageCars", "TotRmsAbvGrd", "BedroomAbvGr", "Fireplaces"]], sqfteq, baths] , axis=1 )
y = df["SalePrice"]
X = sm.add_constant(x)
model = sm.OLS(y,X).fit()
predictions = model.predict(X)

print_model = model.summary()
print(print_model)
```

## 1.4 Remove outliers from your dataset and run the regression again. Do you see a change in some coefficients? Which seem to move the most?

```python
df1 = pd.concat( [ df[["SalePrice",  "OverallQual", "GarageCars", "TotRmsAbvGrd", "BedroomAbvGr", "Fireplaces"]],\
                  sqfteq, baths] , axis=1 )
df1 = df1[(np.abs(stats.zscore(df1)) < 3).all(axis=1)]
df1.columns.values
```

```python
##### Your Code Here #####
x = df1.drop(["SalePrice"],axis =1)
y = df1["SalePrice"]


X = sm.add_constant(x)
model = sm.OLS(y,X).fit()
predictions = model.predict(X)

print_model = model.summary()
print(print_model)
```

## 1.5 Create a new log(y) variable and use it to run a log-linear regression of your variables using statmodels 

```python
##### Your Code Here #####
x = df1.drop(["SalePrice"],axis =1)
y = np.log(df1["SalePrice"])


X = sm.add_constant(x)
model = sm.OLS(y,X).fit()
predictions = model.predict(X)

print_model = model.summary()
print(print_model)
```

## 2.4 Recalculate your regression using Robust Standard Errors? What happens to your standard errors?

```python
##### Your Code Here #####
df2 = df1.drop(labels = ["1F","2F","1FSquared","2FSquared","TotRmsAbvGrd"],axis=1)
x = df2.drop(["SalePrice"],axis=1)
y = np.log(df2["SalePrice"])
X = sm.add_constant(x)
model = sm.OLS(y,X).fit(cov_type="HC3")
predictions = model.predict(X)

print_model = model.summary()
print(print_model)
```
