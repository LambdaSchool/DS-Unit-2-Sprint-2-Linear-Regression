## 1.1 Choose an X and Y variable from your dataset and use them to create a Seaborn Regplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
##### Your Code Here #####
url = 'https://raw.githubusercontent.com/ryanleeallred/datasets/master/Ames%20Housing%20Data/train.csv'

df = pd.read_csv(url)
df.head()
```

```python
x = df['GrLivArea'].values
y = df['SalePrice'].values

fig, ax = plt.subplots(figsize=(8,8))
sns.regplot(x,y,ax=ax);
```

```python
df.isna().sum()
```

```python
df.fillna(method='ffill',inplace=True)
```

## 1.2 Now using the X variables that you feel like will be the best predictors of y use statsmodel to run the multiple regression between these variables and Y. You don't need to use every X variable in your dataset, in fact it's probably better if you don't. Just pick ones that you have already cleaned that seem the most relevant to house prices.

```python
##### Your Code Here #####
df_select = df.select_dtypes(include=['float','int'])
df_select = df_select.drop(columns=['Id'])
df_select.head()
```

```python
df_select = df_select.drop(columns=['GarageYrBlt','MasVnrArea','MiscVal','3SsnPorch', 
                                   'ScreenPorch','EnclosedPorch','MSSubClass',])
df_select.head()
```

## 1.3 Identify the standard errors and P-Values of these coefficients in the output table. What is the interpretation of the P-values here?

```python
##### Your Code Here #####
X = df_select.drop(columns=['SalePrice'])
y = df_select['SalePrice']
X = sm.add_constant(X)

model = sm.OLS(y,X).fit()
predictions = model.predict(X)
print(model.summary())
```

## 1.4 Remove outliers from your dataset and run the regression again. Do you see a change in some coefficients? Which seem to move the most?

```python
##### Your Code Here #####
df_n = df_select[(np.abs(stats.zscore(df_select)) < 3).all(axis=1)]

X = df_n.drop(columns=['SalePrice'])
y = df_n['SalePrice']
X = sm.add_constant(X)

model = sm.OLS(y,X).fit()
preds = model.predict(X)
print(model.summary())
```

## 1.5 Create a new log(y) variable and use it to run a log-linear regression of your variables using statmodels 

```python
##### Your Code Here #####
df_select['ln_price'] = np.log(df_select['SalePrice'])

X = df_select.drop(columns=['SalePrice','ln_price'])
y = df_select['ln_price']
X = sm.add_constant(X)

model = sm.OLS(y,X).fit()
preds = model.predict(X)
print(model.summary())
```

## 2.6 Use seaborn's residplot to plot the distribution of each x variable's residuals. Does these plots indicate any other features that would be potential candidates for polynomial features.

```python
X = df_vif.drop(columns=['SalePrice','ln_price'])
y = df_vif['ln_price']
```

```python
##### Your Code Here #####
sns.set(style='darkgrid')

fig, ax = plt.subplots(ncols=3,nrows=5, figsize=(14,12))
pos = []

for i in range(5):
    for j in range(3):
        pos.append((i,j))

for col,p in zip(X.columns,pos):
    
    sns.residplot(X[col],y,data=df_vif,color='r', ax=ax[p[0]][p[1]],
                  scatter_kws=dict(color='green'),
                  line_kws=dict(color='red'),
                  robust=True)
plt.tight_layout()
```
