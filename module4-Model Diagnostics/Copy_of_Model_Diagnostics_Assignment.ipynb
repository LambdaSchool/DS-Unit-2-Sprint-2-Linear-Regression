```
class Reg_and_Res: 
  '''To pass data to this class, 
    1. do your normalizing and cleaning
    2. make your ones column
    3. ONLY PASS IN A PERFECT SQUARE # OF FEATURES ---- 
        I didn't make it safe yet. 
  
  '''
  def __init__(self, dat, target): 
    assert 'ones' in dat.columns
    assert len(dat.columns)-1 in [x**2 for x in range(32)]
    self.df = dat
    self.m = int(np.floor(np.sqrt(df.shape[1] - 1)))
    self.target = target
    self.crds = self.enumerpairs(self.m)
  
  
  def enumerpairs(self, n): 
    return [(k//n, k%n) for k in range(n**2)]

  def regplots(self): 
  
    fig, axs = plt.subplots(ncols=self.m, nrows=self.m, figsize=(16,12))

    for k in range(self.m**2):
      sns.regplot(self.df[self.df.columns[k]], self.df[self.target], 
                  ax=axs[self.crds[k][0]][self.crds[k][1]])
    pass 
  
  def resplots(self): 
    
    fig, axs = plt.subplots(ncols=self.m, nrows=self.m, figsize=(16,12))
    
    for k in range(self.m**2):
      sns.residplot(self.df[self.df.columns[k]], self.df[self.target], 
                    lowess=True, color="r", 
                    ax=axs[self.crds[k][0]][self.crds[k][1]], 
                    scatter_kws={'color':'blue'})
    pass

MYhat = Reg_and_Res(df, 'lpsa')
MYhat.regplots()
MYhat.resplots()
```
